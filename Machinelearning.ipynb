```python
import re

word = "hi"
pattern = r"h+i+"

# Generate variations using regular expression
variations = []
for i in range(1, len(word) + 4):
    variation = re.sub(pattern, word, "+" * i)
    variations.append(variation)

# Print variations
for variation in variations:
    print(variation)
```

```python
import re

word = "hi"
pattern = r"(hi+|hello)"

# Generate variations using regular expression
variations = []
variations.append(word)  # Add the base word "hi" as the first variation
variations.append("hello")  # Add "hello" as another variation

# Find additional variations matching the pattern
matches = re.findall(pattern, "hii hello hi hiiii")
variations.extend(matches)

# Print variations
for variation in variations:
    print(variation)
```

```python
import csv
from better_profanity import profanity

def remove_abusive_words(csv_file_path, clean_csv_file_path):
    with open(csv_file_path, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)  # Skip the header row
        rows = list(reader)

    cleaned_rows = []
    for row in rows:
        cleaned_row = []
        for value in row:
            # Remove abusive words from each value in the row
            cleaned_value = profanity.censor(value)
            cleaned_row.append(cleaned_value)
        cleaned_rows.append(cleaned_row)

    with open(clean_csv_file_path, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(header)
        writer.writerows(cleaned_rows)

# Example usage
csv_file_path = 'cleaned_output.csv'  # Path to the input CSV file
clean_csv_file_path = 'cleaned_data.csv'  # Path to the cleaned CSV file

remove_abusive_words(csv_file_path, clean_csv_file_path)
```

```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter

# Read the CSV file
data = pd.read_csv("/content/goc_corpus.csv")

# Assuming the target column name is 'target'
X = data.drop('Heading', axis=1)
y = data['Heading']

# Oversampling
oversampler = RandomOverSampler()
X_oversampled, y_oversampled = oversampler.fit_resample(X, y)
print("Oversampled dataset shape:", X_oversampled.shape)
print("Oversampled target counts:", Counter(y_oversampled))

# Undersampling
undersampler = RandomUnderSampler()
X_undersampled, y_undersampled = undersampler.fit_resample(X, y)
print("Undersampled dataset shape:", X_undersampled.shape)
print("Undersampled target counts:", Counter(y_undersampled))
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

data = pd.read_csv("/content/output.csv")

sentences = data['Sentence']
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(sentences)

# Get feature names (words)
feature_names = vectorizer.get_feature_names_out()

# Iterate over each sentence and find 5 most important words
for i in range(len(sentences)):
    sentence = sentences[i]
    tfidf_values = tfidf_matrix[i].toarray()[0]
    top_5_indices = tfidf_values.argsort()[-5:][::-1]
    top_5_words = [feature_names[idx] for idx in top_5_indices]

    print("Sentence:", sentence)
    print("Top 5 important words:", top_5_words)
    print()
```
